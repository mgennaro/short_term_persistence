```python
from astropy.io import fits
from stwcs import updatewcs
from drizzlepac import astrodrizzle
import glob, os, shutil
from multiprocessing import Pool
import matplotlib.pyplot as plt
import numpy as np
from multiprocessing import Pool
%matplotlib inline
```

```python
def driz_single(im):
    flts = sorted(glob.glob('*flt.fits'))
    if os.path.exists(flts[0].replace('flt', 'drz')):
        astrodrizzle.AstroDrizzle(im,final_refimage=flts[0].replace('flt', 'drz'), median=False,blot=False,driz_cr=False, final_wcs=True)
    else:
        astrodrizzle.AstroDrizzle(im, median=False,blot=False,driz_cr=False)
```

```python
for f in glob.glob('*flt.fits'):
    driz_single(f)
```
